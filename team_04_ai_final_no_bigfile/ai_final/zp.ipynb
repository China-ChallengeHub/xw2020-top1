```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
# 支持中文
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号

pd.set_option('display.max_columns', 600)
pd.set_option('display.max_rows', 600)
```

```python
root_path  = 'data/final_data/'
train = pd.read_csv(root_path+'sensor_train_final.csv')
test = pd.read_csv(root_path+'sensor_test_final.csv')
sub = pd.read_csv(root_path+'submit_example.csv')
y = train.groupby('fragment_id')['behavior_id'].min()
test['fragment_id'] += 10000
label = 'behavior_id'
data = pd.concat([train, test], sort=False)
df = data.drop_duplicates(subset=['fragment_id']).reset_index(drop=True)[['fragment_id', 'behavior_id']]
train.head()
sub.head()
```

```python
print(train.shape)
print(test.shape)
print(train["behavior_id"].nunique())
print(train["fragment_id"].nunique())
print(test["fragment_id"].nunique())
```

```python
fps_count = train.groupby(["fragment_id"])["time_point"].count()
import seaborn as sns
plt.figure(figsize=(16,9))
sns.countplot(fps_count)
plt.show()
fps_count = test.groupby(["fragment_id"])["time_point"].count()
import seaborn as sns
plt.figure(figsize=(16,9))
sns.countplot(fps_count)
plt.show()

fps_count = data.groupby(["fragment_id"])["time_point"].count()
import seaborn as sns
plt.figure(figsize=(16,9))
sns.countplot(fps_count)
plt.show()
```

```python
plt.figure(figsize=(16,9))
for i in range(20):
    tmp = train[train.behavior_id==i]
    fps_count = tmp.groupby(["fragment_id"])["time_point"].count()
#     print(fps_count)
    sns.distplot(fps_count,kde=True,label=str(i),hist=False)
plt.show()
```

```python
for i in range(20):
    tmp = train[train.behavior_id==i]
    fps_count = tmp.groupby(["fragment_id"])["time_point"].count()
#     print(fps_count)
    sns.countplot(fps_count)
    plt.show()
plt.show()
```
